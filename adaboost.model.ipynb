```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python
df=pd.read_csv(r'/content/Social_Network_Ads.csv')
```

```python
x = df[['Age','Salary']].values
x.shape
```

```python
y=df['Purchased'].values
y.shape
```

```python
sc = StandardScaler()
x_sc=sc.fit_transform(x)
```

```python
x_sc[:5,:]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_sc,y,test_size=0.2,random_state=100)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
model= AdaBoostClassifier(DecisionTreeClassifier(max_depth=1),n_estimators=25)
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
model = AdaBoostClassifier (estimator = GaussianNB(),n_estimators=6)
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
from sklearn.svm import SVC
model = AdaBoostClassifier (estimator = SVC(),n_estimators=25)
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = AdaBoostClassifier(estimator=RandomForestClassifier(n_estimators=5,max_depth=1),n_estimators=2)
model.fit(x_train,y_train)
model.score(x_test,y_test)
```
